```python
import numpy as np
import pandas as pd
```

# 2018 Movies

```python
link = "https://en.wikipedia.org/wiki/List_of_American_films_of_2018"
df = pd.read_html(link,header=0)[2]
df2 = pd.read_html(link,header=0)[3]
df3 = pd.read_html(link,header=0)[4]
df4 = pd.read_html(link,header=0)[5]
df = df.append(df2.append(df3.append(df4)))
df
```

```python
# Setting Up TMDB 

from tmdbv3api import TMDb
import json
import requests
tmdb = TMDb()
tmdb.api_key = 'e93bb48535e8ca12c047ab10972bd4e8'
```

```python
# Function to get Movie Details From TMDB API

from tmdbv3api import Movie

tmdb_movie = Movie()

def get_movie_details(x):
    genres = []
    result = tmdb_movie.search(x)
    movie_id = result[0].id
    response = requests.get(f'https://api.themoviedb.org/3/movie/{movie_id}?api_key={tmdb.api_key}')
    json_data = response.json()
    
    if json_data['genres']:
        st=" "
        for i in range(0,len(json_data['genres'])):
            genres.append(json_data['genres'][i]['name'])
        return st.join(genres)
    else:
        return np.nan
```

```python
df['genres'] = df['Title'].map(lambda x:get_movie_details(str(x)))
```

```python
df.head(10)
```

```python
movies_2018 = df[['Title','Cast and crew','genres']]
movies_2018.head(10)
```

```python
def get_director_name(x):
    if "(director)" in x:
        return x.split("(director)")[0]
    elif "(directors)" in x:
        return x.split("(directors)")[0]
    else:
        return x.split("(director/screensplay)")[0]
```

```python
movies_2018['director_name'] = movies_2018['Cast and crew'].map(lambda x: get_director_name(x))
```

```python
movies_2018
```

```python
def get_actor_1_name(x):
    return ((x.split("screenplay); ")[-1]).split(", ")[0])
```

```python
movies_2018['actor_1_name'] = movies_2018['Cast and crew'].map(lambda x: get_actor_1_name(x))
```

```python
def get_actor_2_name(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 2:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[1])
```

```python
movies_2018['actor_2_name'] = movies_2018['Cast and crew'].map(lambda x: get_actor_2_name(x))
```

```python
def get_actor_3_name(x):
    if len((x.split("screenplay); ")[-1]).split(", ")) < 3:
        return np.NaN
    else:
        return ((x.split("screenplay); ")[-1]).split(", ")[2])
```

```python
movies_2018['actor_3_name'] = movies_2018['Cast and crew'].map(lambda x: get_actor_3_name(x))
```

```python
movies_2018.head(10)
```

```python
movies_2018 = movies_2018.rename(columns={"Title":"movie_title"})
```

```python
final_2018 = movies_2018[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
final_2018.head(10)
```

```python
final_2018.isnull().sum()
```

```python
final_2018 = final_2018.replace(np.nan,'unknown')
final_2018.isnull().sum()
```

```python
final_2018['comb'] = final_2018['actor_1_name']+' '+final_2018['actor_2_name']+' '+final_2018['actor_3_name']+' '+final_2018['director_name']+' '+final_2018['genres']
final_2018.head(10)
```
