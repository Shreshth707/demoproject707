```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('movie_metadata.csv')
df.head(10)
```

```python
df.columns
```

```python
df = df[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','movie_title']]
df.head(10)
```

```python
df.isnull().sum()
```

```python
df['director_name'] = df['director_name'].fillna('unknown')
df['actor_1_name'] = df['actor_1_name'].fillna('unknown')
df['actor_2_name'] = df['actor_2_name'].fillna('unknown')
df['actor_3_name'] = df['actor_3_name'].fillna('unknown')
```

```python
df.head(10)
```

```python
df['genres'] = df['genres'].str.replace('|', ' ')
```

```python
df.head(10)
```

```python
df['movie_title'] = df['movie_title'].str.lower()
```

```python
df['movie_title'][0]
```

```python
df['movie_title'] = df['movie_title'].apply(lambda x : x[:-1])
```

```python
df['movie_title'][0]
```

```python
df['comb'] = df['actor_1_name']+' '+df['actor_2_name']+' '+df['actor_3_name']+' '+df['director_name']+' '+df['genres']
```

```python
df.head(10)
```

```python
df.to_csv('preprocessing1.csv',index=False)
```
