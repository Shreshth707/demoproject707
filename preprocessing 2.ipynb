```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('credits.csv')
credits.head(10)
```

```python
movie_data = pd.read_csv('movies_metadata.csv')
movie_data.head(10)
```

```python
movie_data.columns
```

```python
movie_data['release_date'] = pd.to_datetime(movie_data['release_date'],errors='coerce')
movie_data['year'] = movie_data['release_date'].dt.year
```

```python
movie_data['year'].value_counts().sort_index()
```

```python
movies_2016_above = movie_data.loc[movie_data['year']>=2017,['genres','id','title','year']]
movies_2016_above
```

```python
movies_2016_above['id'] = movies_2016_above['id'].astype(int)
```

```python
df = pd.merge(movies_2016_above,credits,on='id')
df.head(10)
```

```python
import ast 
df['genres'] = df['genres'].map(lambda x: ast.literal_eval(x))
df['cast'] = df['cast'].map(lambda x: ast.literal_eval(x))
df['crew'] = df['crew'].map(lambda x: ast.literal_eval(x))
```

```python
# Changing Science Fiction genre to Sci-Fi
def create_genrelist(x):
    genres = []
    st=" "
    for genre in x:
        if genre.get('name') == 'Science Fiction':
            genres.append('Sci-Fi')
        else:
            genres.append(genre.get('name'))
    if genres == []:
        return np.nan
    return st.join(genres)
```

```python
df['genres'] = df['genres'].map(lambda x: create_genrelist(x))
df
```

```python
df['crew'][0]
```

```python
def get_director_name(x):
    directors = []
    st = " "
    for director in x:
        if director.get('job') == 'Director':
            directors.append(director.get('name'))
    if directors == []:
        return np.NaN
    else:
        return (st.join(directors))
```

```python
df['director_name'] = df['crew'].map(lambda x: get_director_name(x))
```

```python
def get_actor_1_name(x):
    casts = []
    for cast in x:
        casts.append(cast.get('name'))
    if casts == []:
        return np.nan
    else:
        return casts[0]
```

```python
df['actor_1_name'] = df['cast'].map(lambda x: get_actor_1_name(x))
```

```python
def get_actor_2_name(x):
    casts = []
    for cast in x:
        casts.append(cast.get('name'))
    if casts == [] or len(casts)<2:
        return np.nan
    else:
        return casts[1]
```

```python
df['actor_2_name'] = df['cast'].map(lambda x: get_actor_2_name(x))
```

```python
def get_actor_3_name(x):
    casts = []
    for cast in x:
        casts.append(cast.get('name'))
    if casts == [] or len(casts)<3:
        return np.nan
    else:
        return casts[2]
```

```python
df['actor_3_name'] = df['cast'].map(lambda x: get_actor_3_name(x))
```

```python
df.head(10)
```

```python
final_data = df[['director_name','actor_1_name','actor_2_name','actor_3_name','genres','title']]
final_data.head(10)
```

```python
final_data.isnull().sum()
```

```python
final_data = final_data.replace(np.nan,'unknown')
```

```python
final_data.isnull().sum()
```

```python
final_data = final_data.rename(columns={'title':'movie_title'})
final_data['movie_title'] = final_data['movie_title'].str.lower()
```

```python
final_data['comb'] = final_data['actor_1_name']+' '+final_data['actor_2_name']+' '+final_data['actor_3_name']+' '+final_data['director_name']+' '+final_data['genres']
```
